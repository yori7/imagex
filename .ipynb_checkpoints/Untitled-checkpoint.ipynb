```python
import pandas as pd

elements = []

with open("colortable.html", mode="r") as f:
    contents = f.read()
    
contents = contents.replace("\n", "").replace("</tr>", "").replace("</td>", "").replace("</tbody></table>", "").replace("<b>", "").replace("</b>", "")
contents = [ls.split("<td>") for ls in contents.split("<tr>")]
for content in contents:
    if len(content) == 4:
        element = ["\"" + content[1].split("/")[0] + "\"", content[3].replace("(", "{").replace(")", ",255}")]
        while " \"" in element[0]:
            element[0] = element[0].replace(" \"", "\"")
        element = ": ".join(element)
        elements.append(element)

# df = pd.DataFrame(elements)
elements = "\n".join(elements)
print(elements)
```
